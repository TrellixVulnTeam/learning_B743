````python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for ind1 in range(1,7):
        plt.plot(x, np.sin(x + ind1/2)*(7-ind1)*flip)
````

````python
sinplot() #matplotlib default
````

# figure theme
 * ```darkgrid```: grid helps the plot serve as a lookup table for quantitative information and the white-on grey helps to keep the grid from completing with lines that represent data.
 * ```whitegrid```: better suited to plot with heavy data elements
 * ```dark```: remove grids to proved impressions of patterns in the data
 * ```white```
 * ```ticks```: little exra structure

````python
sns.set_style('darkgrid')
# sns.set_style('dark')
sinplot()
````

````python
sns.set_style('whitegrid')
sns.boxplot(data=np.random.normal(size=(20,6))+np.arange(6)/2)
````

# color palettes
1. reference
 * [Subtleties of Color](https://earthobservatory.nasa.gov/blogs/elegantfigures/2013/08/05/subtleties-of-color-part-1-of-6/)
 * [Diverging Color Maps for Scientific Visualization](https://cfwebprod.sandia.gov/cfdocs/CompResearch/docs/ColorMapsExpanded.pdf)
 * [Choosing Colormaps](https://matplotlib.org/users/colormaps.html)
2. never use ```jet``` colormap
3. general kinds of color palletes
 * Categorical palettes: distinguish discrete chunks of data that do not have an inherent ordering
4. getter and setter
 * ```sns.color_palette()```
 * ```sns.set_palette()```

````python
sns.set(rc={'figure.figsize':(6,6)})
````

## Categorical palette

````python
# default palette
current_palette = sns.color_palette()
sns.palplot(current_palette)
# six variations: deep, muted, pastel, bright, dark, colorblind
````

````python
# circular color systems, use 'hls' color space
sns.palplot(sns.color_palette('hls',6))
````

````python
# select evenly spaced hues while keeping the apprent brightness and satuation much more uniform
sns.palplot(sns.color_palette('husl',6))
````

````python
sns.palplot(sns.hls_palette(8, l=0.3, s=0.8))
````

````python
# another source of visually pleasing categorical palettes
sns.palplot(sns.color_palette('Paired'))
````

````python
sns.palplot(sns.color_palette('Set2',10))
# sns.choose_colorbrewer_palette()
````

````python
sns.set_palette('husl')
sinplot()
````

# sequential color palettes
* never use ```jet```: large hue shifts tend to introduce discontinuities
* for sequential data, better to use palettes that have at most a relatively subtle shift in hue accompanied by a large shift in brightness and satuation.
* ```sns.kdeplot()```
* ```sns.corrplot()```

````python
sns.palplot(sns.color_palette('Blues'))
````

````python
sns.palplot(sns.color_palette('BuGn_r'))
````

````python
sns.palplot(sns.color_palette('GnBu_d'))
````

````python
# sequential cubehelix palette, preserve information when converted to black and white background
sns.palplot(sns.color_palette('cubehelix',8))
````

````python
sns.palplot(sns.cubehelix_palette(8))
````
